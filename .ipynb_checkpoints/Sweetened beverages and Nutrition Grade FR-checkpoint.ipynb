## Sweetened beverages and Nutrition Grade FR

### Abstract
We want to check relation for sugar in Sweetened beverages and Nutrition Grade FR.

We get data from https://world.openfoodfacts.org/cgi/search.pl with filter category "Sweetened beverages".

What is Nutrition Grade FR?
This front-of-pack label classifies foods and beverages according to five categories of nutritional value and quality that are indicated by a color scale ranging from dark green (grade A) to dark red (grade E). Products designated grade A (dark green) have the best nutritional quality while products graded E (dark red) have poor nutritional quality. While the label is voluntary and manufacturers and retailers do not have to use it, it is the only labeling scheme to have official government recognition in France. (https://www.nycfoodpolicy.org/france-becomes-second-country-europe-implement-government-official-front-pack-nutrition-labeling-system/)

We want to prove whether or not the nutrition_grade_fr is correlated with the sugars_100g. Therefore:
* $H_0:$ nutrition_grade_fr is not correlated with the sugars_100g. `d_nutrition_grade_fr_sugars_100g` and `e_nutrition_grade_fr_sugars_100g` are two samples from the same distribution.
* $H_1:$ `d_nutrition_grade_fr_sugars_100g` and `e_nutrition_grade_fr_sugars_100g` come from two different distributions.

```python
sugars_100g_by_nutrition_grade_fr = sugar.groupby("nutrition_grade_fr")["sugars_100g"]
```

```python
d_nutrition_grade_fr_sugars_100g = sugars_100g_by_nutrition_grade_fr.get_group("d")
e_nutrition_grade_fr_sugars_100g = sugars_100g_by_nutrition_grade_fr.get_group("e")
print(d_nutrition_grade_fr_sugars_100g.shape)
print(e_nutrition_grade_fr_sugars_100g.shape)
```

```python
test_result = ttest_ind(d_nutrition_grade_fr_sugars_100g, e_nutrition_grade_fr_sugars_100g, equal_var = False)
print(test_result.pvalue)
if test_result.pvalue <= 0.01:
    print("The differences in sugars_100g are significant. Reject H0.")
else:
    print("There's not enough evidence to reject H0. Don't accept or reject anything else.")
```

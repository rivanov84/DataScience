```python
sugar = mpg_data.sort_values(by=["sugars_100g"], ascending=False)[mpg_data["sugars_100g"] > 0 ][['product_name','generic_name','brands','sugars_100g', "nutrition_grade_fr","quantity","countries"]]
```

```python
sugar = sugar.dropna(subset=["nutrition_grade_fr"]) 
```

```python
sugar_by_nutrition_grade_fr = sugar.groupby("nutrition_grade_fr").size()
sugar_by_nutrition_grade_fr
```

```python
sugar[(sugar["nutrition_grade_fr"] == 'a')|(sugar["nutrition_grade_fr"] == 'b')]
```

```python
plt.figure(figsize = (12, 6))

plt.bar(sugar_by_nutrition_grade_fr.index, sugar_by_nutrition_grade_fr)

plt.title("Bevarage Count per nutrition_grade_fr")
plt.xlabel("nutrition_grade_fr")
plt.ylabel("Bevarage Count")
#plt.xticks(range(min_year, max_year + 1))
# plt.yscale("log")
plt.show()
```
